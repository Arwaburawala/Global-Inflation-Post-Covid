```python
#How inflation affects interest rate
df[['inflation_rate','interest_rate']].corr()
```

```python
yearly_data=df.groupby('year')[['inflation_rate','interest_rate']].mean().reset_index()
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
plt.figure(figsize=(10,5))
sns.lineplot(data=yearly_data, x='year', y='inflation_rate',label='Inflation Rate')
sns.lineplot(data=yearly_data, x='year', y='interest_rate',label='Interest Rate')
plt.title("Yearly Trend: Inflation vs Interest Rate")
plt.xlabel("Year")
plt.ylabel("Rate")
plt.legend()
plt.show()
```

```python
#TOP 7 COUNTRIES WITH HIGHY INFLATION RATE
top7=df.groupby('country',as_index=False)['inflation_rate'].mean()
```

```python
top7=top7.sort_values(by='inflation_rate',ascending=False).head(7)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,5))
sns.barplot(data=top7,x='country',y='inflation_rate')
plt.title("COUNTRIES WITH HIGHEST INFLATION RATE(TOP 7)")
plt.xticks(rotation=45)
plt.xlabel("Country")
plt.ylabel("Inflation Rate")
plt.show()
```

```python
ax=sns.barplot(data=top7,x='country',y='inflation_rate')
for bars in ax.containers: ax.bar_label(bars)
```
